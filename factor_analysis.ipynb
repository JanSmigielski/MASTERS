```python
import pandas as pd
```

```python
summary_metrics = pd.read_csv('results/summary_metrics.csv')
factor_metrics = pd.read_csv('results/factor_metrics.csv')
```

```python
summary_metrics
```

```python
factor_metrics
```

```python
summary_df = summary_metrics.round(2)
factor_df = factor_metrics.round(2)

# Keep only necessary columns from summary
summary_cols = ['factor', 'model', 'Sharpe Ratio', 'Appraisal Ratio', 'Alpha T-stat', 'Skewness']
summary_df = summary_df[summary_cols]

# Prepare per-factor tables
factors = summary_df['factor'].unique()

for factor in factors:
    model_metrics = summary_df[summary_df['factor'] == factor].set_index('model')
    
    # Create row for the raw factor itself
    factor_row = factor_df[factor_df['factor'] == factor]
    if not factor_row.empty:
        row = {
            'Sharpe Ratio': factor_row['factor_sharpe'].values[0],
            'Appraisal Ratio': '-',
            'Alpha T-stat': '-',
            'Skewness': factor_row['factor_skew'].values[0]
        }
        model_metrics.loc['factor'] = row

    # Drop 'factor' column for display
    model_metrics = model_metrics.drop(columns='factor', errors='ignore')
    
    print(f"\n=== {factor} ===")
    display(model_metrics)
```
